# **PUNTO 1** 

Lea los datos del archivo [temperatura.csv.](https://github.com/nicoavilan/Bases-de-datos/blob/main/temperatura.csv) En clase utilizamos estos datos para obtener la siguiente gráfica

image.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/nicoavilan/Bases-de-datos/main/temperatura.csv", index_col=0)
df.head()
```

```python
df=df.sort_values("Temperature").reset_index(drop=True)
```

```python
plt.scatter(df.Temperature, df.Sales, color='r')
plt.plot(df.Temperature, df.Sales)
plt.xlabel('Temperatura')
plt.ylabel('Ventas de helados')
plt.title('Ventas de helados por temperatura')
plt.show()
```

Observe que las temperaturas están en escala Fahrenheit. La ecuación que transforma valores en Fahrenheit a Celsius es $$C=\dfrac{5}{9}(F-32).$$

# **PUNTO DOS**

Genere una nueva columna en el `DataFrame` con temperaturas en Celsius y obtenga una gráfica como la siguiente

image.png

```python
df['Celsius']=(5/9*(df['Temperature']-32)).round(2)
df
```

```python
plt.scatter(df.Celsius, df.Sales, color='r')
plt.plot(df.Celsius, df.Sales)
plt.xlabel('Temperatura (°C)')
plt.ylabel('Ventas de helados')
plt.title('Ventas de helados por temperatura')
plt.show()
```

```python
# Crea la figura y los subplots
fig, axes1 = plt.subplots(figsize=(5, 4), dpi=100)
axes2 = fig.add_axes([0.77, 0.75, 0.08, 0.05])

# Configura el primer subplot
axes1.set_title('Ventas de helados por temperatura')
axes1.set_xlabel('Temperatura (°C)')
axes1.set_ylabel('Ventas de helados')
axes1.scatter(df.Celsius, df.Sales, color='k')
axes1.plot(df.Celsius, df.Sales)

#axes2.tick_params(bottom=False, top=False, left=False, right=False)
axes2.set_xlim([0, 10])
axes2.set_ylim([0, 500])
axes2.set_axis_off() # Quita los ejes coordenados

axes2.annotate("Buenas\n ventas", xy=(8, 510), 
               xytext=(-20, 400), 
               arrowprops=dict(facecolor='red', 
                               edgecolor='red',
                               arrowstyle='->', 
                               linewidth=3.5 
                               ))
plt.show()
```

# **PUNTO TRES**

Lea los datos del archivo [ECG.csv.]() Analice los valores de la columna `‘II’` calculando el máximo, el mínimo, la moda, la media, la mediana y la desviación estándar.


```python
import pandas as pd

url = 'https://github.com/nicoavilan/Bases-de-datos/raw/main/ECG.csv'
df = pd.read_csv(url, index_col=0)

df.head()
```

```python
df["'II'"].describe()
```

```python
df["'II'"]=pd.to_numeric(df["'II'"], downcast="float", errors="coerce")
```

```python
df["'II'"].describe()
```

```python
df["'II'"].max() # Máximo
```

```python
df["'II'"].min() # Mínimo
```

```python
df["'II'"].mode() # Moda
```

```python
df["'II'"].value_counts()[df["'II'"].mode()] # Cuantos valores para que sea la moda
```

```python
df["'II'"].mean() # Media
```

```python
df["'II'"].median() # Mediana
```

```python
df["'II'"].std() # Desviación estandar
```

# **PUNTO CUATRO** 

Obtenga las siguiente gráficas

image.png

```python
max_value = df["'II'"].max()
min_value = df["'II'"].min()
mean_value = df["'II'"].mean()

fig, ax = plt.subplots()

# Graficar la serie de datos de la columna 'II', 
# junto con sus valores maximo, minimo, valor promedio

ax.plot(df["'II'"], color='b',label='ECG')
plt.scatter(df["'II'"].idxmax(), max_value, c='r', label='Máximo')
plt.scatter(df["'II'"].idxmin(), min_value, c='k', label='Mínimo')
ax.axhline(mean_value, color='g', linestyle='--', label='Promedio')

# Agregar las leyendas de los valores máximos, mínimos y promedios
ax.legend(loc='upper right')

# Mostrar el gráfico
plt.show()
```

image.png

[Plot de histograma](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
median_value=df["'II'"].median()
mode_value=df["'II'"].mode()

fig, ax = plt.subplots()
ax.hist(df["'II'"], bins=10)

ax.axvline(mean_value, color='deeppink', linestyle='-', label='Promedio')
ax.axvline(median_value, color='deepskyblue', linestyle='-', label='Mediana')
ax.axvline(-0.118, color='yellow', linestyle='-', label='Moda')

ax.legend(loc='upper right')
plt.show()
```
